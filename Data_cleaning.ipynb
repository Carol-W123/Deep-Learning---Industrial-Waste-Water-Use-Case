```
import pandas as pd
import numpy as np
from pyspark.sql.functions import *
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.tree import DecisionTreeRegressor as DTR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor as RFR
```

```
query = "SELECT * FROM master_data"
#query = "SELECT * FROM feature_data"
a = spark.sql(query)
df_a = a.toPandas()
df_a.head()
```

```
list = df_a.columns.difference(['Sample ID', 'Sample Date', 'Cool Deg Days Flag 15'])
df = df_a.copy()
counter = 0
for item in list:
  df[item] = df[item].str.replace(' ', 'nan')
  df[item] = df[item].str.replace('E', 'nan')
  df[item] = df[item].str.replace('M', 'nan')
  df[item] = df[item].str.replace('T', 'nan')
  df[item] = df[item].str.replace('@13:00\n', '')
  df[item] = df[item].str.replace('@04:00\n', '')
  df[item] = df[item].str.replace('@03:00\n', '')
  df[item] = df[item].str.replace('@3:00\n', '')
  df[item] = df[item].str.replace('@4:00\n', '')
  df[item] = df[item].astype('float')
  counter += 1
df.info(verbose = True)
```

```
query = "SELECT * FROM feature_data"
b = spark.sql(query)
df_b = b.toPandas()
df_b.head()
```

```
df_b['Sample Date'] = pd.to_datetime(df_b['Sample Date'])
df_b.info(verbose=True)
```

```
df = df_b.copy()
target_cols = ['Percent Dead', 'Percent Total Stressed ', 'Percent Dead or Stressed', \
               'Percent Lethargic and Pigmentation (L&P)', 'Pecent Extreme Toxic Shock (S)', \
               'Percent Pigmentation (P)', 'Percent Lethargic (L)', 'Percent Disoriented (D)']
y_data = df_b[target_cols]
df.drop(target_cols, axis=1, inplace=True)
df.drop(['Sample Date'], axis=1, inplace=True)
```

```
y = y_data['Percent Dead']
X = df
# Split data into train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)  
# Scale dataset
scaler = preprocessing.StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
y_train_scale = y_train 
y_test_scale = y_test
```

```
minsam = [1, 2, 5, 8, 10, 15]
maxdep = [4, 5, 8, 10, 12, 14, 15, 16, 20]

r2 = -1
ms_best = minsam[0]
md_best = maxdep[0]

for md in maxdep:
  for ms in minsam:
    regr = RFR(n_estimators = 100, max_depth=md, min_samples_leaf=ms, random_state=40)
    regr.fit(X_train_scale, y_train_scale)
    y_pred = regr.predict(X_test_scale)
    if r2_score(y_test_scale, y_pred) > r2:
      ms_best = ms
      md_best = md
      r2 = r2_score(y_test_scale, y_pred)
regr = RFR(max_depth=md_best, min_samples_leaf=ms_best, random_state=40)
regr.fit(X_train_scale, y_train_scale)
y_pred = regr.predict(X_test_scale)
print("Mean squared error: %.2f"
      % mean_squared_error(y_test_scale, y_pred))
print('Variance score: %.2f' % r2_score(y_test_scale, y_pred))
print(ms_best, md_best)
```
